# Contour plot of the Rosenbrock function

$$
f(x,y) = (a - x)^2 + b(y - x^2)^2
$$
for $a = 1$ and $b = 100$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker, cm 
```

```python
def rosenbrock(x, y, a=1, 
               b=100):
    result = (a - x)**2 + b*(y - x**2)**2
    return result
```

```python
x = np.linspace(-2, 2, 80)
y = np.linspace(-1, 3, 80)

x_grid, y_grid = np.meshgrid(x,y)
z_grid = rosenbrock(x_grid, y_grid)

plt.contour(x_grid, y_grid, z_grid,  locator=ticker.LogLocator()) ### creation of a logarithmic axis locator object
plt.colorbar()
plt.plot()
```

```python
x = np.linspace(0.1, 10, 100)
y = np.log10(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
ax.xaxis.set_major_locator(ticker.LogLocator(base=10.0))
ax.yaxis.set_major_locator(ticker.LogLocator(base = 10.0))

ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
ax.yaxis.set_major_formatter(ticker.ScalarFormatter())
```

```python
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```
